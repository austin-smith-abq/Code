```python
import pandas as pd
import requests
import json

url = "https://coagisweb.cabq.gov/arcgis/rest/services/public/recreation/MapServer/0/query"

querystring = {"where":"1=1","text":"","objectIds":"","time":"","geometry":"","geometryType":"esriGeometryEnvelope","inSR":"","spatialRel":"esriSpatialRelIntersects","relationParam":"","outFields":"*","returnGeometry":"true","maxAllowableOffset":"","geometryPrecision":"","outSR":"","returnIdsOnly":"false","returnCountOnly":"false","orderByFields":"","groupByFieldsForStatistics":"","outStatistics":"","returnZ":"false","returnM":"false","gdbVersion":"","f":"pjson"}

headers = {'user-agent': 'vscode-restclient'}

response = requests.request("GET", url, headers=headers, params=querystring)

park_list = response.json()['features']
parks = pd.json_normalize(park_list)
parks.columns = parks.columns.str.lstrip("attributes.")
```
