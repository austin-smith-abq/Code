```python
import pandas as pd
import json
import requests
token = open(".opensecret").read()
```

```python
url = "http://www.opensecrets.org/api/"

querystring = {"method":"getLegislators","id":"NM","apikey":token}

headers = {'user-agent': 'vscode-restclient'}

response = requests.request("POST", url, headers=headers, params=querystring)
```

```python
legislators = pd.read_xml(response.text)
```

```python
legislators
```

```python
def get_summary(cid):

    url = "http://www.opensecrets.org/api/"

    querystring = {"method":"candSummary","cid":cid,"cycle":"2022","apikey":token}

    headers = {'user-agent': 'vscode-restclient'}

    response = requests.request("POST", url, headers=headers, params=querystring)

    return pd.read_xml(response.text)
```

```python
legislator_summaries = pd.concat([get_summary(i) for i in legislators.cid])
```

```python
legislator_summaries
```

```python
legislator_contributors = pd.concat([get_contributor(i) for i in legislators.cid])
```
